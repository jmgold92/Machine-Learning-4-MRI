```python
%matplotlib inline 
import seaborn as sns
import pandas as pd
iris = sns.load_dataset("iris")
iris.head()
```

```python

# "Melt" the dataset to "long-form" or "tidy" representation
iris_melted = pd.melt(iris, "species", var_name="measurement")
iris_melted.head()
```

```python
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="measurement", y="value", hue="species", split=True,data=pd.melt(iris, "species", var_name="measurement"))
```
